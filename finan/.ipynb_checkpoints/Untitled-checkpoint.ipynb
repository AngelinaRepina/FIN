```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('.xlsx')
data[''] = pd.to_datetime(data[''])
data.set_index('', inplace=True)


plt.figure(figsize=(12,6))
plt.plot(data['  '])  
plt.title('   ')
plt.xlabel('')
plt.ylabel('')
plt.grid()
plt.show()

```

```python
data['Trend'] = data['  '].rolling(window=30).mean()  #     
plt.plot(data['  '], label='')
plt.plot(data['Trend'], label='', color='orange')
plt.title(' ')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.grid()
plt.show()
```

```python
#      -
from statsmodels.tsa.stattools import adfuller

```

```python
residuals = data['  '] - data['Trend'] #
residuals = residuals.dropna()
result = adfuller(residuals)
#print(f'ADF : {result[0]}')
print(f': {result[1]}')
# 
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(residuals)
plt.title(' ')
plt.show()


```

```python
from statsmodels.tsa.arima.model import ARIMA

model_ar1 = ARIMA(data['  '], order=(1,0,0)).fit()
model_ar2 = ARIMA(data['  '], order=(2,0,0)).fit()
model_arma = ARIMA(data['  '], order=(1,0,1)).fit()
```

```python
print(f'AR(1) AIC: {model_ar1.aic}, BIC: {model_ar1.bic}')
print(f'AR(2) AIC: {model_ar2.aic}, BIC: {model_ar2.bic}')
print(f'ARMA(1,1) AIC: {model_arma.aic}, BIC: {model_arma.bic}')
```
