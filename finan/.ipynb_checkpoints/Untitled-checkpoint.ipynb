```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('германия.xlsx')
data['Дата'] = pd.to_datetime(data['Дата'])
data.set_index('Дата', inplace=True)


plt.figure(figsize=(12,6))
plt.plot(data['Курс марки ФРГ'])  
plt.title('Временной ряд цен акций')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.grid()
plt.show()

```

```python
data['Trend'] = data['Курс марки ФРГ'].rolling(window=30).mean()  # Сглаживание с помощью скользящего среднего
plt.plot(data['Курс марки ФРГ'], label='Цена')
plt.plot(data['Trend'], label='Тренд', color='orange')
plt.title('Тренд акций')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.legend()
plt.grid()
plt.show()
```

```python
#Проверка на стационарность с помощью теста Дики-Фуллера
from statsmodels.tsa.stattools import adfuller

```

```python
residuals = data['Курс марки ФРГ'] - data['Trend'] #остатки
residuals = residuals.dropna()
result = adfuller(residuals)
#print(f'ADF статистика: {result[0]}')
print(f'Вероятность: {result[1]}')
#ряд стационарен
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(residuals)
plt.title('Коррелограмма остатков')
plt.show()


```

```python
from statsmodels.tsa.arima.model import ARIMA

model_ar1 = ARIMA(data['Курс марки ФРГ'], order=(1,0,0)).fit()
model_ar2 = ARIMA(data['Курс марки ФРГ'], order=(2,0,0)).fit()
model_arma = ARIMA(data['Курс марки ФРГ'], order=(1,0,1)).fit()
```

```python
print(f'AR(1) AIC: {model_ar1.aic}, BIC: {model_ar1.bic}')
print(f'AR(2) AIC: {model_ar2.aic}, BIC: {model_ar2.bic}')
print(f'ARMA(1,1) AIC: {model_arma.aic}, BIC: {model_arma.bic}')
```
